```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_csv("../data/processed/delivery_data_features.csv")

print(df.shape)
```

```python
top_source_hubs = (
    df.groupby("source_center")
      .size()
      .reset_index(name="shipments")
      .sort_values("shipments", ascending=False)
)

top_source_hubs.head(10)
```

```python
top_destination_hubs = (
    df.groupby("destination_center")
      .size()
      .reset_index(name="shipments")
      .sort_values("shipments", ascending=False)
)

top_destination_hubs.head(10)
```

```python
top_source_hubs.head(20).to_csv(
    "../reports/top_source_hubs.csv",
    index=False
)

top_destination_hubs.head(20).to_csv(
    "../reports/top_destination_hubs.csv",
    index=False
)

print("Hub reports saved.")
```

```python
top_corridors = (
    df.groupby("corridor")
      .size()
      .reset_index(name="shipments")
      .sort_values("shipments", ascending=False)
)

top_corridors.head(10)
```

```python
top_corridors.head(20).to_csv(
    "../reports/top_corridors.csv",
    index=False
)

print("Top corridor report saved.")
```

```python
corridor_delay = (
    df.groupby("corridor")
      .agg({
          "segment_actual_time": "mean",
          "segment_osrm_time": "mean"
      })
      .reset_index()
)

corridor_delay["avg_delay"] = (
    corridor_delay["segment_actual_time"]
    - corridor_delay["segment_osrm_time"]
)

corridor_delay = corridor_delay.sort_values(
    "avg_delay",
    ascending=False
)

corridor_delay.head(10)
```

```python
corridor_delay.head(20).to_csv(
    "../reports/delay_hotspots.csv",
    index=False
)

print("Delay hotspot report saved.")
```

```python
hub_delay = (
    df.groupby("destination_center")
      .agg({
          "segment_actual_time": "mean",
          "segment_osrm_time": "mean"
      })
      .reset_index()
)

hub_delay["avg_delay"] = (
    hub_delay["segment_actual_time"]
    - hub_delay["segment_osrm_time"]
)

hub_delay = hub_delay.sort_values(
    "avg_delay",
    ascending=False
)

hub_delay.head(10)
```

```python
hub_delay.head(20).to_csv(
    "../reports/hub_delay_hotspots.csv",
    index=False
)

print("Hub delay report saved.")
```

```python
route_summary = (
    df.groupby("route_type")
      .agg({
          "segment_actual_time": "mean",
          "segment_osrm_time": "mean",
          "segment_osrm_distance": "mean"
      })
      .reset_index()
)

route_summary["avg_delay"] = (
    route_summary["segment_actual_time"]
    - route_summary["segment_osrm_time"]
)

route_summary
```

```python
route_summary.to_csv(
    "../reports/route_type_summary.csv",
    index=False
)

print("Route summary saved.")
```

```python
bottlenecks = (
    corridor_delay.copy()
)

bottlenecks["bottleneck_score"] = (
    bottlenecks["avg_delay"]
    * bottlenecks["segment_actual_time"]
)

bottlenecks = bottlenecks.sort_values(
    "bottleneck_score",
    ascending=False
)

bottlenecks.head(10)
```

```python
bottlenecks.head(20).to_csv(
    "../reports/bottleneck_corridors.csv",
    index=False
)

print("Bottleneck report saved.")
```
