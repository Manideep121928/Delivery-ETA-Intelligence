## Data Quality Findings

### Issues Detected

1. Negative values in segment_actual_time
2. Negative values in segment_factor
3. Zero values in segment_osrm_time
4. Extreme outliers in delivery duration columns

### Observations

OSRM travel time systematically underestimates actual travel time.

Median delay factor ≈ 1.86x.

```python
df = df[df["segment_actual_time"] >= 0]
```

```python
time_cols = [
    "trip_creation_time",
    "od_start_time",
    "od_end_time",
    "cutoff_timestamp"
]

for col in time_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce", format="mixed")

df[time_cols].head()
```

```python
df[time_cols].isna().sum()
```

```python
df["trip_hour"] = df["trip_creation_time"].dt.hour
df["trip_dayofweek"] = df["trip_creation_time"].dt.dayofweek
df["trip_month"] = df["trip_creation_time"].dt.month
df["trip_weekend"] = (df["trip_dayofweek"] >= 5).astype(int)

df[[
    "trip_hour",
    "trip_dayofweek",
    "trip_month",
    "trip_weekend"
]].head()
```

```python
df["corridor"] = (
    df["source_center"].astype(str)
    + "->"
    + df["destination_center"].astype(str)
)

df["corridor"].head()
```

```python
print("Unique Source Centers:", df["source_center"].nunique())
print("Unique Destination Centers:", df["destination_center"].nunique())
```

```python
all_centers = set(df["source_center"]) | set(df["destination_center"])
print("Total Unique Centers:", len(all_centers))
```

```python
df_clean = df.copy()

# already cleaned negative segment actual time earlier
df_clean = df_clean[df_clean["segment_actual_time"] >= 0].copy()

# drop impossible OSRM zero rows for modeling and analysis
df_clean = df_clean[df_clean["segment_osrm_time"] > 0].copy()

# drop negative factor rows if any remain
df_clean = df_clean[df_clean["segment_factor"] >= 0].copy()

print("After cleaning:", df_clean.shape)
```

```python
df_clean.to_csv("../data/processed/delivery_data_clean.csv", index=False)
```
