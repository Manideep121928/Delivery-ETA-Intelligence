```python
import pandas as pd
import networkx as nx
import numpy as np

pd.set_option("display.max_columns", None)

df = pd.read_csv("../data/processed/delivery_data_features.csv")

print(df.shape)
```

```python
G = nx.from_pandas_edgelist(
    df,
    source="source_center",
    target="destination_center",
    create_using=nx.DiGraph()
)

print("Nodes:", G.number_of_nodes())
print("Edges:", G.number_of_edges())
```

```python
degree_centrality = nx.degree_centrality(G)

degree_df = pd.DataFrame(
    degree_centrality.items(),
    columns=["hub", "degree_centrality"]
)

degree_df = degree_df.sort_values(
    "degree_centrality",
    ascending=False
)

degree_df.head(10)
```

```python
degree_df.head(20).to_csv(
    "../reports/degree_centrality_hubs.csv",
    index=False
)

print("Degree centrality report saved.")
```

```python
betweenness = nx.betweenness_centrality(
    G,
    k=200,
    seed=42
)

betweenness_df = pd.DataFrame(
    betweenness.items(),
    columns=["hub", "betweenness_centrality"]
)

betweenness_df = betweenness_df.sort_values(
    "betweenness_centrality",
    ascending=False
)

betweenness_df.head(10)
```

```python
betweenness_df.head(20).to_csv(
    "../reports/betweenness_centrality_hubs.csv",
    index=False
)

print("Betweenness report saved.")
```

```python
network_summary = pd.DataFrame({
    "Metric": [
        "Nodes",
        "Edges"
    ],
    "Value": [
        G.number_of_nodes(),
        G.number_of_edges()
    ]
})

network_summary
```

```python
network_summary.to_csv(
    "../reports/network_summary.csv",
    index=False
)

print("Network summary saved.")
```
